4.Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.

```python
import csv
#Method  - 1 Using Multiple Line Plot

x = []
y1 = []
y2 = []
y3 = []
y4 = []
with open('fdata.csv','r') as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    next(data,None)
    for row in data:
        x.append(row[0])
        y1.append(row[1])
        y2.append(row[2])
        y3.append(row[3])
        y4.append(row[4])
plt.title("Financial Data Graph" , fontsize = 20)
plt.xlabel("X - axis" ,fontsize = 15)
plt.ylabel("Y - axis" ,fontsize = 15)
plt.plot(x, y1, label = "open")
plt.plot(x, y2, label = "high")
plt.plot(x, y3, label = "low")
plt.plot(x, y4, label = "close")
plt.legend(loc = 4)

# Method - 2 Using Dataframe

df = pd.read_csv("fdata.csv")
print(df.head())

df.plot(x="Date", y=["Open", "High","Low","Close"])
plt.show()

# Method - 3 

df = pd.read_csv('fdata.csv', sep=',', parse_dates=True, index_col=0)
print(df)
df.plot()
plt.show()
```

12.Write a Python program to create multiple types of charts

```python
x = np.arange(0,4*np.pi,0.1)
sine = np.sin(x)
cosine = np.cos(x)
tangent = np.tan(x)
log_range = np.arange(1,10)
log = np.log(log_range)
fig, axs = plt.subplots(2 , 2 ,figsize=(10,10))
fig.suptitle('Multiple charts',fontsize = 20)
fig.tight_layout(pad=5.0)
axs[0,0].plot(x, sine ,'b')
axs[0,0].set_title('Sine',fontsize = 14 )
axs[0,0].set_xlabel('X - Axis')
axs[0,0].set_ylabel('Y - Axis')
axs[0,1].plot(x, cosine ,'r')
axs[0,1].set_title('Cosine',fontsize = 14 )
axs[0,1].set_xlabel('X - Axis')
axs[0,1].set_ylabel('Y - Axis')
axs[1,0].plot(x, tangent ,'g')
axs[1,0].set_title('Tangent',fontsize = 14 )
axs[1,0].set_xlabel('X - Axis')
axs[1,0].set_ylabel('Y - Axis')
axs[1,1].plot(log_range, log ,'y')
axs[1,1].set_title('Log',fontsize = 14 )
axs[1,1].set_xlabel('X - Axis')
axs[1,1].set_ylabel('Y - Axis')
```

13.Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with linestyle -, width .5. and color blue.

```python
df = pd.read_csv("fdata.csv")
x=np.array(list(df["Date"]))
y=np.array(list(df["Close"]))
plt.xlabel("Date" ,fontsize = 15)
plt.ylabel("Closing value " ,fontsize = 15)
plt.plot(x,y,color = 'k')
plt.grid(color = 'blue', linestyle = '-', linewidth = 0.5)
plt.show()
```

14.Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks.

```python
df = pd.read_csv("fdata.csv")
x=np.array(list(df["Date"]))
y=np.array(list(df["Close"]))
plt.xlabel("Date" ,fontsize = 15)
plt.ylabel("Closing value " ,fontsize = 15)
plt.plot(x,y,color = 'k')
plt.grid(which = "major",color = 'lightskyblue', linestyle = '-', linewidth = 2)
plt.minorticks_on()
plt.grid(which = "minor" ,color = 'green', linestyle = ':', linewidth = 1)
plt.tick_params(left=False ,bottom = False) 
plt.show()
```
