1.Write a Python programming to create a pie chart of the popularity of programming Languages.

```python
import matplotlib.pyplot as plt
programming_lan = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
fig, ax = plt.subplots()
plt.pie(popularity, labels=programming_lan)
fig.set_facecolor('white')
```

2.Write a Python programming to create a pie chart with a title of the popularity of programming Languages

```python
programming_lan = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
fig, ax = plt.subplots()
plt.pie(popularity, labels=programming_lan, autopct='%.2f%%')
fig.set_facecolor('white')
plt.title("Popularity of Programming Language")
```

3.Write a Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file.

```python
import pandas as pd
df =  pd.read_csv('medal.csv')
country_data = df["country"]
medal_data = df["gold_medal"]
fig, ax = plt.subplots()
explode = (0.1, 0, 0, 0, 0)  
fig.set_facecolor('white')
plt.pie(medal_data, labels=country_data, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Gold medal achievements of five most successful\ncountries in 2016 Summer Olympics")
plt.show()
```
