5.Write a Python programming to display a bar chart of the popularity of programming Languages. Attach a text label above each bar displaying its popularity (float value).

```python
def addlabels(programming_lan,popularity):
    """ 
        Description: 
            This function is printing value on the top of each bar plot
        Parameter:
            It takes x and y axis label as as argument
        Return:
            None
    """
    for i in range(len(programming_lan)):
        plt.text(i, popularity[i], popularity[i] ,ha = 'center')
        
programming_lan = [ 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.xlabel("Languages",fontsize = 14)
plt.ylabel("Popularity" ,fontsize = 14)
plt.title("Popularity of Programming Language" , fontsize = 18)
plt.bar(programming_lan, popularity ,color = ['indigo','mediumblue','forestgreen','gold','orangered','maroon'])
addlabels(programming_lan,popularity)
plt.show()
```

10.Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

```python
# xlabels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = (22, 30, 35, 35, 26)
women_means =  (25, 32, 30, 35, 29)

fig , ax = plt.subplots()
width = 0.2
index = np.arange(len(men_means))
bar1 = plt.bar(index,men_means,width,color = 'b',label = 'Men')
bar2 = plt.bar(index+width,women_means,width,color = 'hotpink' , label = 'Women')
plt.xlabel("Groups",fontsize = 14)
plt.ylabel("Score" ,fontsize = 14)
plt.title("Scores by group and gender" , fontsize = 18)
x = [ 'Group1', 'Group2', 'Group3', 'Group4', 'Group5' ]
ax.set_xticks([0,1,2,3,4],x)
ax.bar_label(bar1, padding=3)
ax.bar_label(bar2, padding=3)
plt.legend()
plt.show()
```

12.Write a Python program to create bar plots with error bars on the same figure. Sample Date
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
mean_velocity =[ 0.2474, 0.1235, 0.1737, 0.1824 ]
std_dev_of_velocity =[ 0.3314, 0.2278, 0.2836, 0.2645 ]
plt.ylabel('Scores')
plt.xlabel('Velocity')
plt.title('Scores by Velocity')

plt.bar([1,2,3,4],mean_velocity ,yerr = std_dev_of_velocity , color = 'lightskyblue',capsize = 5)
plt.show()
```

13.Write a Python program to create bar plots with errorbars on the same figure. Attach a text label above each bar displaying men means (integer value).

```python
def addlabels(bar_graph):
    """ 
        Description: 
            This function is printing value on the top of each bar plot
        Parameter:
            It takes bar_graph as argument
        Return:
            None
    """
    for bar in bar_graph:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., 2.5*height,'%d' % int(height), ha='center', va='bottom')
        
mean_velocity =[ 0.2474, 0.1235, 0.1737, 0.1824 ]
std_dev_of_velocity =[ 0.3314, 0.2278, 0.2836, 0.2645 ]

fig, ax = plt.subplots()
bar_graph = ax.bar([1,2,3,4],mean_velocity, width = 0.5, color='lightskyblue', yerr=std_dev_of_velocity,capsize = 5)

plt.ylabel('Scores')
plt.xlabel('Velocity')
plt.title('Scores by Velocity')
ax.bar_label(bar_graph , padding = 3)
#addlabels(bar_graph)
plt.show()

```
