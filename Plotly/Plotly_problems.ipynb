1.Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
import plotly.express as px
import numpy as np
x = np.random.randint(1,1000, size=(1000))
y = np.random.randint(1,1000, size=(1000))
px.scatter(x = x, y = y)
```

2.Write a program to draw line and scatter plots for random 100 x and y coordinates

```python
import plotly.express as px
import numpy as np
x = np.random.randint(1,100, size=(100))
y = np.random.randint(1,100, size=(100))
px.line(x=x, y=y)
```

3.Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
import plotly.express as px
import numpy as np
x = np.random.randint(1,1000, size=(1000))
y = np.random.randint(1,1000, size=(1000))
px.scatter(x=x, y=y ,color = y ,size=x)
```

4.Write a program to draw a scatter plot for a given dataset and show datalabels on hover

```python
import pandas as pd
df = pd.read_csv('usa_states.csv')
px.scatter(df, x="State", y="Population", color="Rank" , hover_data=['Postal'])
```
