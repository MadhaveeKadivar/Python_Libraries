1.Write a program to draw box plot of life expectancy by continent for a data set given in a url

```python
import seaborn as sns
import pandas as pd

dataset = pd.read_csv('https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv')

sns.set_style('darkgrid')

# titanic_dataset = sns.load_dataset(titanic)

sns.boxplot(x = dataset["lifeExp"], y = dataset["continent"])
```
