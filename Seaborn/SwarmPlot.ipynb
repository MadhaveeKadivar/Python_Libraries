1.Write a program to draw a swarm plot of total bill against size for a given dataset

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips= pd.read_csv('tips.csv')

sns.set_style('darkgrid')
plt.figure(figsize=(16,9))
plt.xlabel("size",fontsize = 16)
plt.ylabel("total_bill",fontsize = 16)
sns.swarmplot(x = tips["size"], y = tips["total_bill"] ,size = 5)
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(16,9))
plt.xlabel("size",fontsize = 16)
plt.ylabel("total_bill",fontsize = 16)
sns.swarmplot(x = tips["total_bill"], y = tips["day"] ,size = 5)
```
